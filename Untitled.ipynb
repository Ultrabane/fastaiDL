```python
path = '/home/jupyter/tutorials/fastai/nbs/dl1/at1'
data = ImageDataBunch.from_folder(path,ds_tfms=get_transforms(),size=224)
data.normalize(imagenet_stats)
```

```python
data.show_batch(rows=3,figsize=(20,11))
```

```python
learn = ConvLearner(data,models.resnet34,metrics=accuracy)
learn.fit(2)
```

```python
learn.fit(5)
```
